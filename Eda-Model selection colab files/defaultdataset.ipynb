```python
dataframe = df.groupby(['make_model'])['province'].agg(pd.Series.mode).to_frame()

for column in df.columns :
  if column == 'province':
    continue
  if df[f'{column}'].dtype == 'O':  
    dataframe = pd.concat([dataframe,df.groupby(['make_model'])[f'{column}'].agg(pd.Series.mode).to_frame()],axis=1)
  elif df[f'{column}'].dtype == 'float64':
    dataframe = pd.concat([dataframe,df.groupby(['make_model'])[f'{column}'].agg(pd.Series.mean).to_frame()],axis=1)
  elif df[f'{column}'].dtype == 'int64':
    dataframe = pd.concat([dataframe,df.groupby(['make_model'])[f'{column}'].agg(pd.Series.median).to_frame()],axis=1)
  else : 
    dataframe = pd.concat([dataframe,df.groupby(['make_model'])[f'{column}'].agg(pd.Series.mode).to_frame()],axis=1)
```
