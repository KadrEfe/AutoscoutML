```python
dfm = df[['fuel_consumption','extras','mileage','co2_emissions','url','body_type','doors','Mileage','last_service','last_timing_belt_change','\ncolour',' upholstery ','gears']]
```

```python
dfm.columns = dfm.columns.str.strip()
dfm.head()
```

```python
dfm.upholstery = dfm.upholstery.apply(lambda x : re.sub(r'[^A-Za-z]+[^\sA-Za-z]','',str(x)))
dfm.upholstery = dfm.upholstery.str.strip()
dfm.upholstery
m1=dfm.upholstery
m1.value_counts()
```

```python
dfm.colour = dfm.colour.apply(lambda x : re.sub(r'[^A-Za-z]','',str(x)))
dfm.colour.unique()
m2=dfm.colour
```

```python
dfm.gears = dfm.gears.apply(lambda x : re.findall(r'[,]([a-zA-Z0-9]*)[,]',str(x))[0])
dfm.gears.replace({'None': np.nan , 'nan':np.nan},inplace = True)
m3=dfm.gears
```

```python
dfm['gears'].isna().sum()
```

```python
dfm.doors = dfm.doors.apply(lambda x : re.sub(r'[^0-9]','',str(x)))
dfm["doors"] = pd.to_numeric(dfm["doors"])
dfm['doors'] = dfm['doors'].astype('Int64')
m4=dfm.doors
```

```python
dfm['body_type']=dfm['body_type'].str.extract("[,]([\w* \/\-]*)[,]")
m5=dfm.body_type
m5
```

```python
dfm['co2_emissions']=dfm['co2_emissions'].str.extract("[,](\d*)")
m6=dfm.co2_emissions
m6
```

```python
dfm['mileage']=dfm['mileage'].str.replace(r'[^0-9]+', '')
m7=dfm.mileage
```

```python
dfm.extras= dfm.extras.str.replace(r"([\['\]])",'')
dfm.extras= dfm.extras.str.replace('([^\w]\s)',',')
dfm.extras = dfm.extras.str.replace('\(\d{1,3}"\)','').str.strip()
dfm.extras = dfm.extras.str.replace('\"','').str.strip()
dfm.extras = dfm.extras.str.replace(',*[\s],',',')
dfm.extras = dfm.extras.str.replace(',[\s]',',')
m8 = dfm.extras
```

```python
dfm['extras'].str.get_dummies(sep = ",").head()
```

```python
# feature engineering
comb = dfm['fuel_consumption'].str.extract("'(\d\d?.?\d? \d?)")
city = dfm['fuel_consumption'].str.extract("'(\d\d?.?\d? \d?)\w\/[0-9]* [a-z]* \(city\)")
country = dfm['fuel_consumption'].str.extract("\(city\)\'\, \'(\d\d?.?\d?)")
m9=pd.concat([comb, city, country], axis=1)
m9.columns = ['combination(L/100Km)', 'city(L/100Km)', 'country(L/100Km)']
m9
```
